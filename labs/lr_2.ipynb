```python
"""1."""
data['timeStamp'] = pd.to_datetime(data['timeStamp'], format="%Y-%m-%d %H:%M:%S")

data.rename(columns={"lng":"ing", "timeStamp":"accident_time", "twp":"town", "addr":"address"}, inplace=True)
data
```

```python
"""2."""
data = data.drop(['zip', 'desc', 'e', 'address'],  axis=1)
data
```

```python
"""3."""
sort_data = data.sort_values(by=['town', 'lat', 'ing', 'accident_time', 'title'], ascending=False)
sort_data

"""4."""
new_df = sort_data['town'].value_counts(ascending = False).rename_axis('town').reset_index(name='freq')
df=sort_data.merge(new_df,how='inner',on=['town'])
df.sort_values(by=['freq'],ascending=True)
df
```

```python
"""5"""
new_df=new_df.reset_index()
new_df=new_df.set_index('town')
town_list=[x for x in new_df.iloc[0:2].append(new_df.iloc[-2:])['index']]
new_df_1=data[data.town.isin(town_list)]
new_df_1
```

```python
"""6"""
new_df_2=sort_data[(sort_data.town.isin(town_list)==False) & (sort_data['town'].isnull()==False)]
hour=[x.hour for x in pd.to_datetime(new_df_2['accident_time'],format="%Y-%m-%d %H:%M:%S")]
new_df_2.insert(5, "hour", hour, True)
new_df_2
```

```python
"""7"""
new_df_3 = new_df_2['hour'].value_counts(ascending = False).rename_axis('hour').reset_index(name='freq')
df_1=new_df_2.merge(new_df_3,how='inner',on=['hour'])
df_1.sort_values(by=['freq'],ascending=False)
df_1
```

```python
"""8"""
from sklearn.preprocessing import minmax_scale
df_1['freq'] = minmax_scale(df_1['freq'])
df_1
```
